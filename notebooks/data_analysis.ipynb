# Data Analysis of Parentalidade Survey

This notebook contains the analysis of the dataset collected from the parentalidade survey. The analysis includes data exploration, visualization, and insights derived from the survey data.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')
```

```python
# Load the dataset
data_path = '../data/raw/survey_data.csv'
survey_data = pd.read_csv(data_path)

# Display the first few rows of the dataset
survey_data.head()
```

```python
# Data exploration
# Summary statistics
survey_data.describe()
```

```python
# Visualizations
# Plotting the distribution of responses
plt.figure(figsize=(10, 6))
sns.histplot(survey_data['Qual idade dos seus filhos?'], bins=10, kde=True)
plt.title('Distribution of Children Ages')
plt.xlabel('Age of Children')
plt.ylabel('Frequency')
plt.show()
```
