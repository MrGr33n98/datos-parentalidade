# Data Analysis on Parentalidade Dataset

This notebook performs data analysis on the parentalidade dataset. It includes loading the data, performing exploratory data analysis (EDA), and visualizing the results.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data_path = '../data/processed/cleaned_data.csv'
df = pd.read_csv(data_path)

# Display the first few rows of the dataset
df.head()
```

```python
# Exploratory Data Analysis (EDA)

# Summary statistics
df.describe()

# Check for missing values
df.isnull().sum()
```

```python
# Visualizations

# Distribution of a specific column
plt.figure(figsize=(10, 6))
sns.histplot(df['column_name'], bins=30, kde=True)
plt.title('Distribution of Column Name')
plt.xlabel('Column Name')
plt.ylabel('Frequency')
plt.show()
```
